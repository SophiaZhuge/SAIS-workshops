## Numpy
- NumPy, which stands for Numerical Python is a Python library used for scientific computing.
- It can also be used as an efficient multi-dimensional container for data and and a collection of routines for processing those arrays.
- It is the most basic and a powerful package for working with data in python.

More practice materials:    [bit.ly/python_numpy](bit.ly/python_numpy)


```python
# usually we add the second part 'as np' so we can access it with 'np.command' instead of 'numpy.command'
import numpy as np 
```

```python
# Return a new array of given shape and type, without initializing entries. Default type is float.
L = np.empty(3, dtype=int)
for i in range(3):
    L[i] = (i+5)/2
print(L)
```

```python
# Sum the Values in an Array
s = np.sum(L)
print(s)
```

```python
# Return the non-negative square-root of an array, element-wise.
sq = np.sqrt(L)
print(sq)
```

```python
# Natural logarithm, element-wise.
l = np.log(L)
print(l)
```

```python
# Calculate the exponential of all elements in the input array.
L_1 = np.exp(l)
print(L_1)
```

## Scikit-learn.preprocessing
- The sklearn.preprocessing package provides several common utility functions and transformer classes to change raw feature vectors into a representation that is more suitable for the downstream estimators.

- In general, learning algorithms benefit from standardization of the data set. If some outliers are present in the set, robust scalers or transformers are more appropriate.

More details:  [scikit-learn.preprocessing](https://scikit-learn.org/stable/modules/preprocessing.html#preprocessing)

```python
# show data
titanic_data
```

```python
# drop useless columns, i.e. PassengerId, Name, Ticket
# (axis = 1) means column data
titanic_data = titanic_data.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
```

```python
# show data
titanic_data
```

```python
# data information
titanic_data.info()
```

```python
# convert int to double
for col in ['Survived', 'Pclass', 'SibSp', 'Parch']:
    titanic_data[col] = titanic_data[col].astype('float64')
```

```python
# data information
titanic_data.info()
```

```python
# define lists with the names of the attribute
attributes_cat = ['Sex', 'Cabin', 'Embarked']
attributes_num = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
```

```python
# check how much NaN value each numerical attribute has
for col in ['Survived'] + attributes_num + attributes_cat:
    print("{} = {}, ".format(col, (pd.isna(titanic_data[col])).sum()/titanic_data[col].count()))
```

```python
# replace NaN with the average age in the Age attribute
m = titanic_data['Age'].mean()
print('Average age: ', m)
def nan_to_zero(data,col):
     data[col] = np.where(pd.isna(data[col]), m, data[col])
# call function
nan_to_zero(titanic_data, 'Age')
# display result
titanic_data['Age'].head(10)
```

```python
# replace NaN with Unknown in category attibutes
def nan_to_unknown(data,col):
     data[col] = np.where(pd.isna(data[col]), 'Unknown', data[col])
# call function
for col in attributes_cat:
    nan_to_unknown(titanic_data, col)
# display result
titanic_data.head(10)
```

```python
# The train_test_split function in scikit-learn allows to easily get these partitions.
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(titanic_data, test_size=0.15, random_state=42)
```

The train and test sets are chosen randomly from all the available data. By specifying a value for random_state, we are making sure that every time we run this instruction, the train and test set will have the exact same instances. random_state "controls the shuffling applied to the data before applying the split".

OneHotEncoder() and StandardScaler() are part of the scikit-learn preprocessing module.

- OneHotEncoder() allows to transform a categorical variable to a one-hot encoding representation.

- StandardScaler() performs feature scaling by standardisation.

- ColumnTransformer() an estimator available in scikit-learn that allows to group different transformations into a single method

```python
# import OneHotEncoder, StandardScaler and ColumnTransformer and create the actual transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

transform = ColumnTransformer([
    ("num", StandardScaler(), attributes_num),
    ("cat", OneHotEncoder(), attributes_cat),
])
```

```python
# separate the target feature from the attributes
train_set_attributes = train_set.drop('Survived', axis=1)
train_set_labels = train_set['Survived']
```

```python
# apply the fit and apply the full transformation to the training data using fit_transform
train_set_attributes_prepared = transform.fit_transform(train_set_attributes)
```
